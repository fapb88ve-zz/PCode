```python
#ESTE ES EL ALGORITMO.

def pic_mod():
    lp = os.listdir(os.getcwd())
    plogo = Image.open('pancita4.png')
    plogow, plogoh = plogo.size
    plogo = plogo.resize((int(plogow*.1), int(plogoh*.1)))
    plogow, plogoh = plogo.size
    for i in lp:
        if '.JPG' in i:
            pic = Image.open(i)
            picw, pich = pic.size
            pic.paste(plogo, (picw - plogow, pich - plogoh), plogo)
            pic.save('Pancita '+ i)
        else:
            pass
pic_mod()
```
